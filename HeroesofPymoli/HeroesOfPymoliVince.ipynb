```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
data_file_pd = pd.read_csv(file_to_load)
data_file_pd
```

## Player Count

* Display the total number of players


```python
total_number_of_players = (data_file_pd["SN"].nunique())
pd.DataFrame({"Total Players": [total_number_of_players]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Number of Unique Items
number_of_unique_items = data_file_pd["Item ID"].nunique()
#Average Purchase Price
average_purchase_price = data_file_pd["Price"].mean()
average_purchase_price = round(average_purchase_price, 2)
#Total Number of Purchases
total_number_of_purchases = data_file_pd["Purchase ID"].count()
#Total Revenue
total_revenue = data_file_pd["Price"].sum()
pd.options.display.float_format = '${:,.2f}'.format
pd.DataFrame({"Number of Unique Items" : [number_of_unique_items],
              "Average Purchase Price" : [average_purchase_price],
              "Total Number of Purchases" : [total_number_of_purchases],
              "Total Revenue" : [total_revenue]})
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Gender Demographics
#Male Count
data_file_male = data_file_pd.loc[data_file_pd['Gender'] == 'Male']
data_file_male_count = data_file_male['SN'].nunique()
#Female Count
data_file_female = data_file_pd.loc[data_file_pd['Gender'] == 'Female']
data_file_female_count = data_file_female['SN'].nunique()
#Other Count
data_file_other = data_file_pd.loc[data_file_pd['Gender'] == 'Other / Non-Disclosed']
data_file_other_count = data_file_other['SN'].nunique()
#Percentage of Female Players
percentage_female_players = (data_file_female_count/total_number_of_players ) * 100
#Percentage of Male Players
percentage_male_players = (data_file_male_count/total_number_of_players ) * 100
#Percentage of Other Players
pd.options.display.float_format = '{:,.2f}'.format
percentage_other_players = (data_file_other_count/total_number_of_players ) * 100
pd.DataFrame({"": ["Male","Female","Other"],"Total Count":[data_file_male_count,data_file_female_count,data_file_other_count],
             "Percentage of Players": [percentage_male_players,percentage_female_players,percentage_other_players]
             })
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Purchasing Analysis(Gender)
#Male Purchase Count
male_purchase_count = data_file_male['Purchase ID'].count()
#Female Purchase Count
female_purchase_count = data_file_female['Purchase ID'].count()
#Other Purchase Count
other_purchase_count = data_file_other['Purchase ID'].count()
#Male Average Purchase
male_average_purchase = data_file_male['Price'].mean()
#Female Average Purchase
female_average_purchase = data_file_female['Price'].mean()
#Other Average Purchase
other_average_purchase = data_file_other['Price'].mean()
#Male Total Purchase
male_total_purchase = data_file_male['Price'].sum()
#Female Total Purchase
female_total_purchase = data_file_female['Price'].sum()
#Other Total Purchase
other_total_purchase = data_file_other['Price'].sum()
#Male Avg Total Purchase per Person
avg_total_male = (male_total_purchase/data_file_male_count)
#Female Avg Total Purchase per Person
avg_total_female = (female_total_purchase/data_file_female_count)
#Other Avg Total Purchase per Person
avg_total_other = (other_total_purchase/data_file_other_count)
pd.options.display.float_format = '${:,.2f}'.format
pd.DataFrame({"Gender" : ["Female","Male","Other/Non-Disclosed"], "Purchase Count":[female_purchase_count,male_purchase_count,other_purchase_count],
             "Average Purchase Price":[female_average_purchase,male_average_purchase,other_average_purchase],
              "Total Purchase Value":[female_total_purchase,male_total_purchase,other_total_purchase],
              "Average Total Purchase Per Person":[avg_total_female,avg_total_male,avg_total_other]
             })
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Age Demographics
#Creating bins and labels
bins = [0,9,14,19,24,29,34,39,150]

age_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
pd.cut(data_file_pd["Age"],bins,labels=age_labels)
#adding age range field to data frame
data_file_pd["Age Range"] = pd.cut(data_file_pd["Age"],bins,labels=age_labels)
data_file_pd
```

```python
data_file_pd_unique = data_file_pd.drop_duplicates('SN')
data_file_pd_unique = data_file_pd_unique.groupby('Age Range').count()['SN'].reset_index()
percentage_count = (data_file_pd_unique['SN'] / data_file_pd_unique['SN'].sum())*100
data_file_pd_unique["Percentage of Players"] = percentage_count
pd.options.display.float_format = '{:,.2f}%'.format
data_file_pd_unique
#pd.cut(data_file_pd_unique["Age"],bins,age_range_labels).value_counts()
#data_file_pd_unique
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Purchasing Analysis (Age)
purchasing_analysis_age = data_file_pd.groupby('Age Range')
purchase_count = (purchasing_analysis_age['Price'].count())
avg_purchase_price = (purchasing_analysis_age['Price'].mean())
total_purchase_value = (purchasing_analysis_age['Price'].sum())
purchasing_analysis_out = pd.merge(purchase_count,avg_purchase_price, on='Age Range',how='inner')
purchasing_analysis_out = pd.merge(purchasing_analysis_out,total_purchase_value, on='Age Range',how='inner')
purchasing_analysis_out = pd.merge(purchasing_analysis_out,data_file_pd_unique, on ='Age Range', how = 'inner') 
purchasing_analysis_out['Avg Total Purchase Per Person'] = purchasing_analysis_out['Price']/ purchasing_analysis_out['SN']
purchasing_analysis_out = purchasing_analysis_out[['Age Range','Price_x','Price_y','Price','Avg Total Purchase Per Person']]
purchasing_analysis_out = purchasing_analysis_out.rename(columns={'Price_x': 'Purchase Count','Price_y':'Average Purchase Price','Price':'Total Purchase Value'})
pd.options.display.float_format = '${:,.2f}'.format
purchasing_analysis_out
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Identify the top 5 spenders
spenders = data_file_pd.groupby('SN')
top_spenders = (spenders['Price'].sum())
top_spenders_count = (spenders['Price'].count())
top_spenders_avg = (spenders['Price'].mean())
#top_spenders_df = pd.DataFrame(top_spenders).sort_values("Price",ascending=False)
top_spenders_df = pd.DataFrame(top_spenders)
top_spenders_df["Purchase Count"] = top_spenders_count
top_spenders_df["Average Purchase Price"] = top_spenders_avg
top_spenders_df = top_spenders_df.sort_values("Price",ascending=False)
top_spenders_df_out = top_spenders_df[['Purchase Count','Average Purchase Price', 'Price']]
top_spenders_df_final = top_spenders_df_out.iloc[:5,:]
top_spenders_df_final = top_spenders_df_final.rename(columns={'Price':'Total Purchase Value'})
top_spenders_df_final
pd.options.display.float_format = '${:,.2f}'.format
spenders.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Most Popular Items by Purchase Count
items = data_file_pd.groupby(['Item ID','Item Name','Price'])
items_purchase_count = items['SN'].count()
item_total_purchase = items['Price'].sum()
#top_items_df = pd.DataFrame(items_purchase_count).sort_values("SN",ascending=False)
top_items_df = pd.DataFrame(items_purchase_count)
top_items_df["Total Purchase Value"] = item_total_purchase
top_items_df_out = top_items_df.rename(columns={'SN':'Purchase Count'})
top_items_df_out = top_items_df_out.sort_values("Purchase Count",ascending=False)
top_items_df_out=top_items_df_out.iloc[:5,:]
pd.options.display.float_format = '${:,.2f}'.format
top_items_df_out
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Most Popular Items by Purchase Value
purchase_value = data_file_pd.groupby(['Item ID'])
items_purchase_total = purchase_value['Price'].sum()
items_purchase_count = purchase_value['SN'].count()
top_totalpurchase_df = pd.DataFrame(purchase_value)
top_totalpurchase_df['Total Purchase Value'] = items_purchase_total
top_totalpurchase_df['Purchase_Count']= items_purchase_count
top_totalpurchase_df.sort_values("Total Purchase Value",ascending=False)
top_totalpurchase_df_out=top_items_df_out.iloc[:5,:]
pd.options.display.float_format = '${:,.2f}'.format
top_totalpurchase_df_out
```
